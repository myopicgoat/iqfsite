```python

# inline function to compute the lagged-log returns for an arbitrary lag
lr = lambda prices, lag: np.log(prices[range(0+lag,len(prices),lag)] / 
                                prices[range(0,len(prices)-lag,lag)])
# compute the log-returns corresponding to ratios over 5 quotes
# play with the lag afterwards to see what changes in the analysis
lag     = 5  
lr_aapl = lr(aapl,lag)
lr_goog = lr(goog,lag)

x    = np.linspace(-.15,.15,500)

lr = lr_aapl
mu,sigma = norm.fit(lr)
yaapl = norm.pdf(x,mu,sigma)


lr = lr_goog
mu,sigma = norm.fit(lr)
ygoog = norm.pdf(x,mu,sigma)

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.hist(lr_aapl, 50, normed=1, facecolor='royalblue', alpha=0.75);
ax2.hist(lr_goog, 50, normed=1, facecolor='orange', alpha=0.75);
ax1.axes.set_xlim([-0.15,0.15])
ax2.axes.set_xlim([-0.15,0.15])

f.tight_layout()

ax1.plot(x,yaapl,color="green",lw=2.5)
ax2.plot(x,ygoog,color="crimson",lw=2.5)

f.savefig("aapl_goog_histlrnormtfit_lagged.svg")
```

```python
lr = lambda prices, lag: np.log(prices[range(0+lag,len(prices),lag)] / 
                                prices[range(0,len(prices)-lag,lag)])

lags     = np.array([1,3,5,7,10,15,20])
sig_aapl = np.zeros(len(lags))
mu_aapl  = np.zeros(len(lags))
sig_goog = np.zeros(len(lags))
mu_goog  = np.zeros(len(lags))
i        = 0
for lag in lags:
    mu_aapl[i],sig_aapl[i] = norm.fit(lr(aapl,lag))
    mu_goog[i],sig_goog[i] = norm.fit(lr(goog,lag))
    i+=1

plt.plot(lags,mu_aapl,marker='o')
plt.plot(lags,mu_goog,marker='o')
plt.legend(["AAPL","GOOG"],fontsize=14,loc=2)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Lag [days]",fontsize=14)
plt.ylabel("Fitted means",fontsize=14)

plt.savefig("aapl_goog_evolmean.svg")

plt.figure()
plt.plot(lags,sig_aapl**2,marker='o')
plt.plot(lags,sig_goog**2,marker='o')
plt.legend(["AAPL","GOOG"],fontsize=14,loc=2)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Lag [days]",fontsize=14)
plt.ylabel("Fitted variances",fontsize=14)

plt.savefig("aapl_goog_evolvar.svg")
```

```python
def callDeltaBS(St,K,tau,r,v):
    d1 = ( (np.log(St/K)) + (r+np.power(v,2)/2)*tau ) / (v*np.sqrt(tau))
    return norm.cdf(d1)

def callPriceBS(St,K,tau,r,v):
    d1 = ( (np.log(St/K)) + (r+np.power(v,2)/2)*tau ) / (v*np.sqrt(tau))
    d2 = d1-v*np.sqrt(tau)
    return St*norm.cdf(d1)-K*np.exp(-r*tau)*norm.cdf(d2)

rangeSt = np.linspace(40,105,100)
K       = 75
tau     = 2
r       = 0.05
v       = 0.1

callDelta  = callDeltaBS(rangeSt,K,tau,r,v)
callPrices = callPriceBS(rangeSt,K,tau,r,v)

f,ax1 = plt.subplots()

l1=ax1.plot(rangeSt,callDelta,lw=2.5,label="Delta")

ax2 = ax1.twinx()
l2=ax2.plot(rangeSt,callPrices,lw=2.5,color="orange",label="Price")

ax1.set_ylim([0,callDeltaBS(rangeSt[-1],K,tau,r,v)])
ax2.set_ylim([0,callPriceBS(rangeSt[-1],K,tau,r,v)])
ax1.set_xlim([40,rangeSt[-1]])

ax2.grid(b=False)

lns = l1+l2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=2,fontsize=14)

ax1.set_xlabel("Price of the underlying",fontsize=14)
ax1.set_ylabel("Delta",fontsize=14)
ax2.set_ylabel("Price of the call", fontsize=14)

f.tight_layout()

f.savefig("deltaCall.svg")

# plt.figure()
# plt.plot(rangeSt,callDelta,lw=2.5)
# plt.plot(rangeSt,callPrices,lw=2.5)
```

```python
def callVegaBS(St,K,tau,r,v):
    d1 = ( (np.log(St/K)) + (r+np.power(v,2)/2)*tau ) / (v*np.sqrt(tau))
    d2 = d1-v*np.sqrt(tau)
    return np.sqrt(tau)*St*norm.pdf(d1)

def callVegaNumerical(St,K,tau,r,v): 
    deltaVol = v/5000
    Vplus    = callPriceBS(St,K,tau,r,v+deltaVol)
    V        = callPriceBS(St,K,tau,r,v)
    return (Vplus-V)/deltaVol

rangeVol = np.linspace(0.001,6.1,500)
St       = 60
K        = 75
tau      = 2
r        = 0.05
v        = 0.1

callVega   = callVegaNumerical(St,K,tau,r,rangeVol)
callPrices = callPriceBS(St,K,tau,r,rangeVol)

f,ax1 = plt.subplots()

l1=ax1.plot(rangeVol,callVega+0.1,lw=2.5,label="Vega")

ax2 = ax1.twinx()
l2=ax2.plot(rangeVol,callPrices,lw=2.5,color="orange",label="Price")

ax1.set_ylim([0,35])
ax2.set_ylim([0,60.2])
ax1.set_xlim([0,6])

ax2.grid(b=False)

lns = l1+l2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=4,fontsize=14)

ax1.set_xlabel("Volatility",fontsize=14)
ax1.set_ylabel("Vega",fontsize=14)
ax2.set_ylabel("Price of the call", fontsize=14)

f.tight_layout()

f.savefig("vegaCall.svg",transparent=True)
```

```python
def callThetaNumerical(St,K,tau,r,v): 
    deltaTau = tau/1000
    Vplus    = callPriceBS(St,K,tau+deltaTau,r,v)
    V        = callPriceBS(St,K,tau,r,v)
    return (Vplus-V)/deltaTau

rangeTau = np.linspace(1,5.1,100)
St       = 60
K        = 75
r        = 0.05
v        = 0.1

callTheta  = callThetaNumerical(St,K,rangeTau,r,v)
callPrices = callPriceBS(St,K,rangeTau,r,v)

f,ax1 = plt.subplots()

l1=ax1.plot(rangeTau,callTheta,lw=2.5,label="Theta")

ax2 = ax1.twinx()
l2=ax2.plot(rangeTau,callPrices,lw=2.5,color="orange",label="Price")

# ax1.set_ylim([0,35])
ax2.set_ylim([0,6.3])
ax1.set_xlim([1,5.1])

ax2.grid(b=False)

lns = l1+l2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=4,fontsize=14)

ax1.set_xlabel("Lifetime of the contract",fontsize=14)
ax1.set_ylabel("Theta",fontsize=14)
ax2.set_ylabel("Price of the call", fontsize=14)

f.tight_layout()

f.savefig("thetaCall.svg")
```

```python

endProbas  = lambda p,N:   [comb(N,k)*(p**k)*(1-p)**(N-k) for k in range(0,N+1)]
endReturns = lambda u,d,N: [(u**k)*(d)**(N-k) for k in range(0,N+1)]

# enter some arbitrary values
p = 0.6  # probability of going up
u = 1.03 # multiplicative "up factor"
d = 0.99 # multiplicative "down factor"

# number of steps (play with this number)
N = 100

returns = endReturns(u,d,N)
probas  = endProbas(p,N)

plt.plot(returns,np.array(probas)+0.001,ls='none',marker='o')
plt.xticks(fontsize=12)
plt.yticks([round(e,2) for e in np.linspace(0,0.08,4)],fontsize=12)

plt.xlabel("Returns",fontsize=14)
plt.ylabel("Probability",fontsize=14)

plt.savefig("binomLN.svg")
```

```python
from numpy import trapz

logreturns = np.log(returns)

th_mean = N*(p*np.log(u)+(1-p)*np.log(d))
th_var  = N*p*(1-p)*np.log(d/u)**2
pr_norm  = probas / abs(trapz(logreturns,probas))
xx = np.linspace(np.min(logreturns),np.max(logreturns),100)
yy = norm.pdf(xx,th_mean,np.sqrt(th_var))

plt.plot(logreturns,pr_norm+0.01,ls='none',marker='o')
plt.plot(xx,yy+0.01)

plt.xlim([th_mean-1.0,th_mean+1.0])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Log-returns",fontsize=14)
plt.ylabel("Probability",fontsize=14)

plt.legend(["Log-returns","Normal fit"],fontsize=14)

plt.savefig("binomNorm.svg")
```
